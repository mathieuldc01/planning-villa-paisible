```python
# Installer la bibliothèque qrcode si ce n'est pas déjà fait
# pip install qrcode[pil]

import qrcode

# Lien à encoder
url = "https://team-dore-evasion.assoconnect.com"

# Génération du QR code
qr = qrcode.QRCode(
    version=1,  # taille du QR code (1 = 21x21 modules)
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,  # taille de chaque "box" en pixels
    border=4,  # largeur de la bordure en boxes
)
qr.add_data(url)
qr.make(fit=True)

# Création de l'image
img = qr.make_image(fill_color="black", back_color="white")

# Sauvegarde de l'image
img.save("C:\\Users\\mathi\\OneDrive\\Bureau\\team_dore_qr_code.png")

print("QR code généré et sauvegardé sous 'qrcode.png'.")
```
